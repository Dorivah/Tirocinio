```python
import numpy as np 
import pandas as pd 
import os
import cv2

import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
```

```python
batch_size = 40
img_height = 256
img_width = 256
```

```python
training_ds = tf.keras.preprocessing.image_dataset_from_directory(
    'C:/Users/lucal/Documents/Tirocinio/dataset_new/train',
    #validation_split=0.2,
    #subset= "training",
    seed=42,
    image_size= (img_height, img_width),
    batch_size=batch_size

)
```

```python
testing_ds = tf.keras.preprocessing.image_dataset_from_directory(
'C:/Users/lucal/Documents/Tirocinio/dataset_new/test',
    #validation_split=0.2,
    #subset= "validation",
    seed=42,
    image_size= (img_height, img_width),
    batch_size=batch_size

)
```

```python
class_names = training_ds.class_names
class_names
```

```python
plt.figure(figsize=(10, 10))
for images, labels in training_ds.take(1):    # il numero tra parentesi indica il gruppo di immagini (batch)
  for i in range(12):
    ax = plt.subplot(3,4 , i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.grid(True)
```

```python
## Ottimizzazione del dataset 
AUTOTUNE = tf.data.experimental.AUTOTUNE
training_ds = training_ds.cache().prefetch(buffer_size=AUTOTUNE)
testing_ds = testing_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
## lets define our CNN
# definisco l'architettura di una rete neurale convoluzionale
MyCnn = tf.keras.models.Sequential([
  layers.experimental.preprocessing.Rescaling(1./255),   # per ridimensionare i valori dei pixel dell'immagine in un intervallo compreso tra 0 e 1 dividendo ogni valore dei pixel per 255
  layers.Conv2D(32, 3, activation='relu'),   # strato di convoluzione 2D con 32 filtri, ciascuno di dimensione 3x3 (estrazione caratteristiche)
  layers.MaxPooling2D(),   # sottocampionamento (pooling) che esegue il max-pooling (riduce la dimensione spaziale dei dati)
  layers.Conv2D(64, 3, activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(128, 3, activation='relu'),
  layers.MaxPooling2D(),

  layers.GlobalAveragePooling2D(),   # riduce le dimensioni spaziali dell'immagine a una singola dimensione (average pooling globale)
  layers.Dense(256, activation='relu'),
  layers.Dense(len(class_names), activation= 'softmax')
])
```

```python
MyCnn.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# configura il processo di addestramento per il modello di rete neurale definito
```

```python
# addestramento della rete neurale (CNN)
retVal = MyCnn.fit(training_ds, validation_data= testing_ds, epochs = 100)
```

```python
#epochs = list(range(1,101))
#plt.plot(epochs, retVal.history['loss'], label = 'training loss')
#plt.plot(epochs, retVal.history['accuracy'], label = 'training accuracy')
#plt.grid(True)
#plt.xticks(epochs)
#plt.legend()

x_values = np.arange(0, 101, 10)

# Imposta i valori personalizzati dell'asse x
plt.xticks(x_values)

# Ora puoi disegnare il grafico
plt.plot(retVal.history['loss'], label='training loss')
plt.plot(retVal.history['accuracy'], label='training accuracy')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# saving model
MyCnn.save('drowsiness-stage1.keras')
```

```python
plt.figure(figsize=(20, 20))
for images, labels in testing_ds.take(2):
    predictions = MyCnn.predict(images)
    predlabel = []
    
    for mem in predictions:
        predlabel.append(class_names[np.argmax(mem)])
    
    for i in range(40):
        ax = plt.subplot(10, 4, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title('Predicted label:'+ predlabel[i])
        plt.axis('off')
        plt.grid(True)
```
